```python
import pandas as pd
import geopandas as gpd
import os

# Define paths
airbnb_path = '../data/raw/airbnb_listings.csv.gz'
zillow_path = '../data/raw/zillow_home_values.csv'
geo_path = '../data/raw/nyc_zip_codes.geojson'

# 1. Check Airbnb
if os.path.exists(airbnb_path):
    df_airbnb = pd.read_csv(airbnb_path, compression='gzip', low_memory=False)
    print(f"✅ Airbnb Data Loaded: {df_airbnb.shape[0]} listings")
else:
    print("❌ Airbnb file not found!")

# 2. Check Zillow
if os.path.exists(zillow_path):
    df_zillow = pd.read_csv(zillow_path)
    print(f"✅ Zillow Data Loaded: {df_zillow.shape[0]} zip codes")
else:
    print("❌ Zillow file not found!")

# 3. Check GeoJSON
if os.path.exists(geo_path):
    gdf_nyc = gpd.read_file(geo_path)
    print(f"✅ NYC Map Loaded: {gdf_nyc.shape[0]} shapes")
else:
    print("❌ GeoJSON file not found!")
```
